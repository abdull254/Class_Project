```python
!pip install pytrends
from pytrends.request import TrendReq
import pandas as pd
import time
from pathlib import Path
```

```python
# 1) CONNECT
# tz=360 means US Central time in pytrends examples

pytrends = TrendReq(hl="en-US", tz=360)
```

```python
# 2) CITIES
cities = {
    # Host cities
    "New York": "US-NY-501",
    "Los Angeles": "US-CA-803",
    "Dallas": "US-TX-623",
    "Houston": "US-TX-618",
    "Atlanta": "US-GA-524",
    "Miami": "US-FL-528",
    "Seattle": "US-WA-819",
    "San Francisco": "US-CA-807",
    "Boston": "US-MA-506",
    "Philadelphia": "US-PA-504",
    "Kansas City": "US-MO-616",

    # Non-host controls
    "Chicago": "US-IL-602",
    "Minneapolis": "US-MN-613",
    "Phoenix": "US-AZ-753",
    "Denver": "US-CO-751",
    "Detroit": "US-MI-505",
    "San Diego": "US-CA-825",
}

keywords = [
    "soccer",
    "youth soccer",
    "MLS",
    "soccer club"
]
```

```python
# 4) TIMEFRAME
timeframe = "2010-01-01 2026-12-31"

# 5) OUTPUT FOLDER
out_dir = Path("google_trends_output")
out_dir.mkdir(exist_ok=True)
```

```python
# 6) FUNCTION TO PULL INTEREST OVER TIME
def fetch_interest_over_time(keyword: str, city_name: str, geo_code: str) -> pd.DataFrame:
# weekly Google Trends
    pytrends.build_payload(
        kw_list=[keyword],
        cat=0,
        timeframe=timeframe,
        geo=geo_code,
        gprop=""
    )

    df = pytrends.interest_over_time()

    if df.empty:
        return pd.DataFrame(columns=["date", "keyword", "city", "geo", "interest"])

    # pytrends returns the keyword column + isPartial
    value_col = keyword

    df = df.reset_index()

    if "isPartial" in df.columns:
        df = df.drop(columns=["isPartial"])

    df = df.rename(columns={value_col: "interest", "date": "date"})
    df["keyword"] = keyword
    df["city"] = city_name
    df["geo"] = geo_code

    return df[["date", "keyword", "city", "geo", "interest"]]
```

```python
# 7) LOOP THROUGH ALL CITY-TERM PAIRS
all_data = []
failed = []

for city_name, geo_code in cities.items():
    for keyword in keywords:
        try:
            print(f"Pulling: {city_name} | {keyword}")
            df = fetch_interest_over_time(keyword, city_name, geo_code)
            all_data.append(df)
            time.sleep(2)  # helps avoid rate limits
        except Exception as e:
            failed.append({
                "city": city_name,
                "geo": geo_code,
                "keyword": keyword,
                "error": str(e)
            })
            time.sleep(5)
```

```python
# 8) COMBINE AND SAVE
if all_data:
    trends_df = pd.concat(all_data, ignore_index=True)
else:
    trends_df = pd.DataFrame(columns=["date", "keyword", "city", "geo", "interest"])

trends_df.to_csv(out_dir / "trends_city_keyword_weekly.csv", index=False)

failed_df = pd.DataFrame(failed)
failed_df.to_csv(out_dir / "failed_requests.csv", index=False)

print("\nDone.")
print(f"Saved main file: {out_dir / 'trends_city_keyword_weekly.csv'}")
print(f"Saved failures:  {out_dir / 'failed_requests.csv'}")
print("\nPreview:")
print(trends_df.head())
```

```python
import pandas as pd

df = pd.read_csv("google_trends_output/trends_city_keyword_weekly.csv")
df["date"] = pd.to_datetime(df["date"])

host_cities = {
    "New York", "Los Angeles", "Dallas", "Houston", "Atlanta",
    "Miami", "Seattle", "San Francisco", "Boston",
    "Philadelphia", "Kansas City"
}

df["host_city"] = df["city"].isin(host_cities).astype(int)
df["year"] = df["date"].dt.year
df["post_2026"] = (df["date"] >= "2026-06-01").astype(int)  # adjust if you want tournament start

# yearly city-level averages
yearly = (
    df.groupby(["city", "host_city", "keyword", "year"], as_index=False)["interest"]
      .mean()
      .rename(columns={"interest": "avg_interest"})
)

yearly.to_csv("google_trends_output/trends_city_keyword_yearly.csv", index=False)
print(yearly.head())
```

```python
soccer_df = yearly[yearly["keyword"] == "soccer"].copy()
soccer_df["post_2026"] = (soccer_df["year"] >= 2026).astype(int)
soccer_df["treat_post"] = soccer_df["host_city"] * soccer_df["post_2026"]

soccer_df.to_csv("google_trends_output/did_soccer_yearly.csv", index=False)
print(soccer_df.head())
```
